```python
# Load the dataset
file_path = "/mnt/data/Enhanced_Synthetic_Mining_Data.csv"
df = pd.read_csv(file_path)

# Set general plot style
sns.set_style("whitegrid")

# Figure 1: Bar Chart - Accident Frequency by Risk Level
plt.figure(figsize=(8, 5))
sns.countplot(x="Job_Risk_Level", hue="Accident_Occurred", data=df, palette="viridis")
plt.title("Accident Frequency by Job Risk Level")
plt.xlabel("Job Risk Level")
plt.ylabel("Count")
plt.legend(title="Accident Occurred", labels=["No", "Yes"])
plt.show()

# Figure 2: Histogram - Distribution of Weather Risk Index
plt.figure(figsize=(8, 5))
sns.histplot(df["Weather_Risk_Index"], bins=30, kde=True, color="royalblue")
plt.title("Distribution of Weather Risk Index")
plt.xlabel("Weather Risk Index")
plt.ylabel("Frequency")
plt.show()

# Figure 3: Heatmap - Correlation between Features
plt.figure(figsize=(10, 6))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# Figure 4: Scatter Plot - Accident Occurrence vs. Machine Age
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Machine_Age_Years", y="Accident_Occurred", data=df, alpha=0.5)
plt.title("Accident Occurrence vs. Machine Age")
plt.xlabel("Machine Age (Years)")
plt.ylabel("Accident Occurred")
plt.show()

# Figure 5: Boxplot - Noise Levels by Shift Type
plt.figure(figsize=(8, 5))
sns.boxplot(x="Shift_Type", y="Noise_Level_dB", data=df, palette="coolwarm")
plt.title("Noise Levels by Shift Type")
plt.xlabel("Shift Type")
plt.ylabel("Noise Level (dB)")
plt.show()
```
